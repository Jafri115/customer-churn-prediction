```python
# api_test.py
import requests
import json

def test_api():
    url = "http://127.0.0.1:54326/predict"
    
    # Test data based on the example in the API schema
    payload = {
        "customers": [
            {
                "gender": "Female",
                "SeniorCitizen": 0,
                "Partner": "Yes",
                "Dependents": "No",
                "tenure": 1,
                "PhoneService": "No",
                "MultipleLines": "No phone service",
                "InternetService": "DSL",
                "OnlineSecurity": "No",
                "OnlineBackup": "Yes",
                "DeviceProtection": "No",
                "TechSupport": "No",
                "StreamingTV": "No",
                "StreamingMovies": "No",
                "Contract": "Month-to-month",
                "PaperlessBilling": "Yes",
                "PaymentMethod": "Electronic check",
                "MonthlyCharges": 29.85,
                "TotalCharges": "29.85"
            }
        ]
    }
    
    try:
        # First check API health
        health_response = requests.get("http://127.0.0.1:8000/health")
        print("API Health Status:", health_response.json())
        
        # Make prediction
        print("Sending prediction request...")
        response = requests.post(url, json=payload)
        
        # Check if successful
        if response.status_code == 200:
            result = response.json()
            print("Prediction successful:")
            print(json.dumps(result, indent=2))
            
            # Print each prediction in a readable format
            for i, prediction in enumerate(result):
                print(f"\nCustomer {i+1}:")
                print(f"  Churn Probability: {prediction['churn_probability']:.2%}")
                print(f"  Churn Prediction: {prediction['churn_prediction']} ({prediction['churn_label']})")
                
        else:
            print(f"Error: Status code {response.status_code}")
            print(response.json())
            
    except requests.exceptions.ConnectionError:
        print("Connection error: Unable to connect to the API.")
        print("Make sure the API server is running at http://127.0.0.1:8000")
    except Exception as e:
        print(f"Error: {str(e)}")

if __name__ == "__main__":
    test_api()
```
